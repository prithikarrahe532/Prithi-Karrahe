```python
from sklearn.preprocessing import LabelEncoder

education_encoder = LabelEncoder()
df['education_encoded'] = education_encoder.fit_transform(df['Education_Level'])
df[['Education_Level', 'education_encoded']].head(10)
```

```python
risk_encoder = LabelEncoder()
df['risk_encoded'] = risk_encoder.fit_transform(df['Risk_Category'])
df[['Risk_Category', 'risk_encoded']].head(10)
```

```python
JOB_encoder = LabelEncoder()
df['JOB_encoded'] = JOB_encoder.fit_transform(df['Job_Title'])
df[['Job_Title', 'JOB_encoded']].head(10)
```

```python
numerical_features = ['Average_Salary', 'Years_Experience', 'AI_Exposure_Index', 'Tech_Growth_Factor', 'Automation_Probability_2030', 'education_encoded', 'risk_encoded','JOB_encoded']
corr_matrix = df[numerical_features].corr()
corr_matrix
```
